```python
import matplotlib.pyplot as plt
x = np.linspace(-3, 3, num=50)
y = 2 * x - 1
z = x ** 3

figure = plt.figure()
axes = figure.add_subplot()
axes.plot(x, y, color="red", linewidth=1.0, linestyle = '-')
axes.plot(x, z, color="blue", linewidth=1.0, linestyle = '-')
```

```python
t = np.arange(0, 2.5, 0.1)
y1 = np.sin(t * np.pi)
y2 = np.cos(t * np.pi)

figure = plt.figure()
axes = figure.add_subplot()
axes.plot(t, y1, "b*")
axes.plot(t, y2, "r^")
plt.show()
```
